```python
import os
import csv
bugdet_data = os.path.join("budget_data.csv")

total_months = 0
net_total = 0
previous_month = 0
average_changes = []
with open(bugdet_data, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader)
    for i in csvreader:
        # The total number of months included in the dataset
        total_months = total_months + 1
        net_total = net_total + int(i[1])
    
        changes = int(i[1]) - previous_month
        average_changes.append(changes)
        previous_month = int(i[1])
        average_list = sum(average_changes)/len(average_changes)
        print(changes)
        print(previous_month)
    print(average_changes)
    print(average_list)

               
```
